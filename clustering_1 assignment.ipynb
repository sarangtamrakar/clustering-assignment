```python
import numpy as np
from sklearn.cluster import KMeans
from scipy import misc
import matplotlib.pyplot as plt
```

```python
# load the dataset

features = misc.face(gray=True)
features.shape
```

```python
row = features.shape[0]
col = features.shape[1]

image = features
image.shape
```

```python
# plot the image before compression
plt.imshow(image,cmap=plt.cm.gray)
```

```python
# feature preprocessing before apply clustering & then apply Kmean clutering because we know the no. of cluster which we want to create
image = features.reshape(-1,1)
agg = KMeans(n_clusters=5)
agg.fit(image)
```

```python
# change the shape of label
label = agg.labels_.reshape(features.shape[0],features.shape[1])
label.shape
```

```python
plt.imshow(label,cmap=plt.cm.gray)
```
